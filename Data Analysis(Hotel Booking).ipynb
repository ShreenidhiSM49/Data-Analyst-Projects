# Data analysis and Visualizations

```python
cancelled_pr=df["is_canceled"].value_counts(normalize=True)
cancelled_pr

plt.figure(figsize=(5,4))
plt.title("Reservation Status Count")
plt.bar(["Not_cancelled","Cancelled"],df["is_canceled"].value_counts())
```

```python
plt.figure(figsize=(8,4))
ax1=sns.countplot(x="hotel",hue="is_canceled",data=df)
plt.title("Reservations status of different hotels",size=25)
plt.xlabel("hotel")
plt.ylabel("Number of reservations")
plt.legend(["not_canceled","cancelled"])
```

```python
resort_hotel=df[df["hotel"]=="Resort Hotel"]
resort_hotel["is_canceled"].value_counts(normalize=True)
```

```python
city_hotel=df[df["hotel"]=="City Hotel"]
city_hotel["is_canceled"].value_counts(normalize=True)
```

```python
resort_hotel=resort_hotel.groupby("reservation_status_date")[["average daily rate"]].mean()
city_hotel=city_hotel.groupby("reservation_status_date")[["average daily rate"]].mean()
```

```python
plt.figure(figsize=(20,8))
plt.title("ADR in City and Resort Hotel",size=(20))
plt.plot(resort_hotel.index,resort_hotel["average daily rate"],label="Resort Hotel")
plt.plot(city_hotel.index,city_hotel["average daily rate"],label="City Hotel")
plt.legend(fontsize=30)


```

```python
df["month"]=df["reservation_status_date"].dt.month
plt.figure(figsize=(16,8))
ax1=sns.countplot(x="month",data=df,hue="is_canceled")
plt.title("Reservation status per month",size=25)
plt.xlabel("Month",size=25)
plt.ylabel("num of reservtaions",size=25)
```

```python
plt.figure(figsize=(16,8))
plt.title("ADR per month",size=20)
sns.barplot("month","average daily rate",data=df[df["is_canceled"]==1].groupby("month")[["average daily rate"]].sum().reset_index())
```

```python
canceled_data=df[df["is_canceled"]==1]
top_10_nations=canceled_data["country"].value_counts()[:10]
plt.figure(figsize=(8,8))
plt.title("Top 10 Countries")
plt.pie(top_10_nations,autopct="%.2f",labels=top_10_nations.index)
```

```python
df["market_segment"].value_counts()
```

```python
df["market_segment"].value_counts(normalize=True)
```

```python
canceled_data["market_segment"].value_counts(normalize=True)
```

```python
canceled_df_adr=canceled_data.groupby("reservation_status_date")[["average daily rate"]].mean()
canceled_df_adr.reset_index(inplace=True)
canceled_df_adr.sort_values("reservation_status_date",inplace=True)

non_canceled_data=df[df["is_canceled"]==0]
non_canceled_df_adr=non_canceled_data.groupby("reservation_status_date")[["average daily rate"]].mean()
non_canceled_df_adr.reset_index(inplace=True)
non_canceled_df_adr.sort_values("reservation_status_date",inplace=True)
plt.figure(figsize=(16,8))
plt.plot(non_canceled_df_adr["reservation_status_date"],non_canceled_df_adr["average daily rate"],label="not cancelled")
plt.plot(canceled_df_adr["reservation_status_date"],canceled_df_adr["average daily rate"],label="cancelled")
plt.legend()
```
